```python
from sklearn.model_selection import train_test_split
df = df.drop(['Time'],axis=1)

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
y.head()
```

DUMMY CLASSIFIER

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import recall_score, accuracy_score, precision_score
from sklearn.metrics import confusion_matrix
dummy_majority = DummyClassifier(strategy = 'most_frequent').fit(X_train, y_train)
y_dummy_predictions = dummy_majority.predict(X_test)
t = (accuracy_score(y_test,y_dummy_predictions),recall_score(y_test,y_dummy_predictions))
t
```

SVC

```python

from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
a = accuracy_score(y_test,y_pred) 
r = recall_score(y_test,y_pred)
p = precision_score(y_test,y_pred)
t = (a,r,p)
t
```

```python
cnf_mat = confusion_matrix(y_test, y_pred.round())
sns.heatmap(cnf_mat, xticklabels = ['Valid','Fraud'],  
            yticklabels = ['Valid','Fraud'], annot = True, fmt ="d"); 
plt.title("Confusion matrix") 
plt.ylabel('True class') 
plt.xlabel('Predicted class')
```
